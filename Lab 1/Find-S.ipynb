```python
import pandas as pd
```

```python
df = pd.read_csv('dataset.csv')
```

```python
cols = df.columns
```

```python
cols
```

```python
h = list()
for i in range(len(cols)-1):
    h.append(None)
```

```python
print(f"H0: {h}")
for i in range(len(df)):
    if df.iloc[i]['Enjoy'] == 'yes':
        for j in range(len(cols)-1):
            col = cols[j]
            if h[j] is None:
                h[j] = df.iloc[i][col]
            elif h[j] != df.iloc[i][col]:
                h[j] = '?'
    print(f"H{i+1}: {h}")
```

```python
print("Specific Hypothesis: ", h)
```
