```python
import numpy as np, pandas as pd, matplotlib.pyplot as plt
```

```python
def getCoefficients(x, y):
    n = len(x)
    x2 = [i**2 for i in x]
    xy = [x[i] * y[i] for i in range(n)]
    b0 = (sum(y) * sum(x2) - sum(x) * sum(xy)) / (n * sum(x2) - sum(x)**2)
    b1 = (n * sum(xy) - sum(x) * sum(y)) / (n * sum(x2) - sum(x)**2)
    return (b0, b1)
```

```python
def getPrediction(x, b):
    return b[0] + b[1] * x
```

```python
def plotRegLine(x, y, b):
    pred = getPrediction(x, b)
    # Plot
    plt.scatter(x, y, color = 'm', marker = 'o', s = 30)
    plt.plot(x, pred, color = 'g')
    plt.xlabel = 'x'
    plt.ylabel = 'y'
    plt.show()
```

```python
dataset = pd.read_csv('salary_data.csv')
```

```python
x = dataset['YearsExperience']
y = dataset['Salary']
```

```python
b = getCoefficients(x, y)
```
