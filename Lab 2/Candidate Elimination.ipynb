```python
df = pd.read_csv('dataset.csv')
df.head()
```

```python
n, m = df.shape[1], df.shape[0]
```

```python
values = list()
for col in df.columns[:-1]:
    values.append(list(set(df[col])))
values
```

```python
def checkConsistency(h1, h2, n):
    i = 0
    while i < n:
        if h1[i] == '?' or h2[i] == '?':
            pass
        elif h1[i] != h2[i]:
            return False
        i += 1
    return True
```

```python
G = [['?' for _ in range(n-1)]]
S = [None for _ in range(n-1)]
for i in range(m):
    if df.iloc[i][-1].lower() == 'yes':
        x = 0
        l = len(G)
        while x < l:
            g = G[x]
            increment = True
            if not checkConsistency(df.iloc[i][0:-1], g, n-1):
                G.remove(g)
                l -= 1
                increment = False
            if increment:
                x += 1
        for j in range(n-1):
            if S[j] is None:
                S[j]=df.iloc[i][j]
            elif S[j] != df.iloc[i][j]:
                S[j] = '?'
    else:
        x = 0
        l = len(G)
        while x < l:
            g = G[x]
            if checkConsistency(df.iloc[i][0:-1], g, n-1):
                for j in range(n-1):
                    if g[j] == '?':
                        g1 = g.copy()
                        if len(values[j]) == 1:
                            continue
                        if values[j][0] == df.iloc[i][j]:
                            g1[j] = values[j][1]
                        else:
                            g1[j] = values[j][0]
                        G.append(g1)
                G.remove(g)
                l -= 1
            else:
                x += 1
    print(f"G{i}: ", G)
    print(f"S{i}: ", S)
    print()
print(f"G: ", G)
```

```python
middle = list()
for g in G:
    for i in range(n-1):
        if g[i] != S[i]:
            x = g.copy()
            x[i] = S[i]
            if x not in middle:
                middle.append(x)
print(f"Middle Layer: ", middle)
```
